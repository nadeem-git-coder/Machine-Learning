```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pie Chart

```python
areas =['MArketing','Sales','Development','HR','Customer Support']
budget =[2.5,4,10,1,6]
```

```python
plt.pie(budget,labels=areas)
plt.show()
```

```python
 # by percentage
```

```python
plt.pie(budget,labels=areas,autopct='%0.1f%%')
plt.show()
```

# Scatter 


```python
x=[23,45,12,49,97,32,11,44]
y =[45,34,67,87,55,22,89,90]

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs Salary Graph')

```

```python
product_id=[1,2,3,4,5,6,7,8]
online_price =[233,456,770,120,222,444,200,300]
offline_price =[400,300,100,333,444,565,565,899]
chor_bazaar = [100,200,300,400,150,250,450,500]
```

```python
plt.plot(product_id , online_price,color='red')
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparison')
plt.show()
```

```python
plt.plot(product_id , online_price,color='red',marker='D',markersize=10)
plt.xlabel('Product')
plt.ylabel('Prices')
plt.title('Product Price Comparison')
plt.show()
```

```python
plt.plot(product_id , online_price,color='red',marker='D',markersize=10,label ='Online Price')
plt.plot(product_id,offline_price,color='yellow',marker=10,label = 'Offline Price')
plt.plot(product_id,chor_bazaar,color='green',marker=10,label = 'Chor Price')
plt.xlabel('Product')

plt.ylabel('Prices')
plt.title('Product Price Comparison')
plt.legend()
plt.show()
```

# Bar Graph

```python
year =[2012,2013,2014,2015,2016,2017,2018]
price =[65000,69000,75000,55000,52000,64000,81000]
plt.bar(year,price)
plt.xlabel('Year')

plt.ylabel('Price')
plt.title('Year by Year Price Comparison')
plt.show()
```
